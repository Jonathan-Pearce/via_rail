```python
import sqlite3
```

```python
con = sqlite3.connect("via_data_test.db")
cur = con.cursor()
```

```python
print(con.execute("SELECT * FROM tbl_via_data_v4").fetchall())

con.execute("""
CREATE VIEW view_via_data_v13 as
SELECT * FROM tbl_via_data_v4 A
LEFT JOIN (
    select * from tbl_stop_v5 B
    LEFT JOIN tbl_location_v2 C
    ON B.ROUTE_STOP = C.LOCATION_ID
    LEFT JOIN tbl_location_v2 D
    ON B.ROUTE_PREV_STOP = D.LOCATION_ID
    LEFT JOIN (
            select * from tbl_train_v4 E
            LEFT JOIN tbl_route_v3 F
            ON E.ROUTE = F.ROUTE_ID
    ) G
    ON B.TRAIN = G.TRAIN_ID
) H
ON A.TRAIN_STOP = H.STOP_ID;""")

print(con.execute("SELECT * FROM view_via_data_v13").fetchall())
```
